```

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to simulate the oscillateur harmonique
def oscillateur_harmonique(m, k, x0, v0):
    # Time array
    t = np.linspace(0, 20, 500)
    
    # Angular frequency
    omega = np.sqrt(k / m)
    
    # Position as a function of time
    x_t = x0 * np.cos(omega * t) + (v0 / omega) * np.sin(omega * t)
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(t, x_t, label=f'm={m}, k={k}, x0={x0}, v0={v0}')
    plt.xlabel('Temps (s)')
    plt.ylabel('Position (m)')
    plt.title('Oscillateur Harmonique')
    plt.legend()
    plt.grid(True)
    plt.show()

# Interactive sliders
m_slider = FloatSlider(min=0.1, max=10, step=0.1, value=1, description='Masse (kg)')
k_slider = FloatSlider(min=0.1, max=10, step=0.1, value=1, description='Raideur (N/m)')
x0_slider = FloatSlider(min=-10, max=10, step=0.1, value=1, description='Position initiale (m)')
v0_slider = FloatSlider(min=-10, max=10, step=0.1, value=0, description='Vitesse initiale (m/s)')

# Create the interactive plot
interact(oscillateur_harmonique, m=m_slider, k=k_slider, x0=x0_slider, v0=v0_slider)
    
```
